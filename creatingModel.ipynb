```python
# %pip install numpy
# %pip install pandas
# %pip install scikit-learn
# %pip install -U imbalanced-learn
# %pip install xgboost
import numpy as np
import pandas as pd

df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv")
df.head()

feature = df.drop(columns="Diabetes_binary").columns
label = "Diabetes_binary"
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size = 0.3, stratify = df.Diabetes_binary, random_state = 1123)
test, validation = train_test_split(test, test_size = 0.3, stratify = test.Diabetes_binary, random_state = 1002)

len(train), len(test), len(validation)
```

```python
train["Diabetes_binary"].value_counts()
```

```python
test["Diabetes_binary"].value_counts()
```

```python
validation["Diabetes_binary"].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler

resampler = RandomUnderSampler(random_state=1909)
train[feature], train[label] = resampler.fit_resample(train[feature], train[label])
train.dropna(inplace=True)

len(train)
train
```

```python
train["Diabetes_binary"].value_counts()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.linear_model import LogisticRegression

def logisticPipeline():
    return make_pipeline(
        make_column_transformer(
            # (StandardScaler(), ["BMI"]),
            (KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["BMI"]),
            (KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["Age"]),
            (KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["MentHlth"]),
            (KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["PhysHlth"]),
            (KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["Income"]),
            remainder="passthrough"
        ),
        LogisticRegression()
    )
```

```python
logistic = logisticPipeline()
logistic
```

```python
logistic.fit(train[feature], train[label])
```

```python
logistic.score(train[feature], train[label])
```

```python
logistic.score(validation[feature], validation[label])
```

```python
from sklearn.inspection import permutation_importance

importance = permutation_importance(logistic, validation[feature], validation[label])
importance = pd.Series(importance.importances_mean, index=feature)
importance.sort_values(ascending=False)
```

```python
logisticFeature = ["GenHlth", "BMI", "Age", "Fruits", "Smoker", "NoDocbcCost"]
```

```python
def logisticPipelineImportance():
    return make_pipeline(
        make_column_transformer(
            # (StandardScaler(), ["BMI"]),
            (KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["BMI"]),
            (KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["Age"]),
            #(KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["MentHlth"]),
            #(KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["PhysHlth"]),
            #(KBinsDiscretizer(encode="ordinal", strategy="quantile"), ["Income"]),
            remainder="passthrough"
        ),
        LogisticRegression()
    )
featureImPipeline = logisticPipelineImportance()
featureImPipeline
```

```python
featureImPipeline.fit(train[logisticFeature], train[label])
```

```python
featureImPipeline.score(train[logisticFeature], train[label])
```

```python
featureImPipeline.score(validation[logisticFeature], validation[label])
```

```python
featureImPipeline.get_params()
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_distributions = {
    'columntransformer__kbinsdiscretizer-1__strategy': ['uniform', 'quantile', 'kmeans'],
     'columntransformer__kbinsdiscretizer-1__encode': ['ordinal', 'onehot', 'onehot-dense'],
     'columntransformer__kbinsdiscretizer-1__n_bins': [5, 10, 20],
    'columntransformer__kbinsdiscretizer-2__strategy': ['uniform', 'quantile', 'kmeans'],
     'columntransformer__kbinsdiscretizer-2__encode': ['ordinal', 'onehot', 'onehot-dense'],
     'columntransformer__kbinsdiscretizer-2__n_bins': [5, 10, 20],
    'logisticregression__C': np.logspace(-4, 4, 20),  # ค่าพารามิเตอร์ C
    'logisticregression__penalty': ['l1', 'l2', 'elasticnet', 'none'],  # ประเภทของการ regularization
    'logisticregression__solver': ['newton-cg', 'lbfgs', 'liblinear', 'saga'],  # วิธีการหาค่าของพารามิเตอร์
    'logisticregression__max_iter': [100, 200, 300, 400, 500],  # จำนวนรอบสูงสุด
    'logisticregression__class_weight': [None, 'balanced']  # การปรับน้ำหนักของคลาส
}

searchLogistic = RandomizedSearchCV(featureImPipeline, param_distributions,  n_iter=100, cv=5, random_state=42)
searchLogistic.fit(train[logisticFeature], train[label])
searchLogistic.best_params_, f"{searchLogistic.best_score_:.3f}"
```

```python
from sklearn.metrics import classification_report

y_pred = searchLogistic.best_estimator_.predict(validation[logisticFeature])
print(classification_report(validation[label], y_pred))
```

```python
from sklearn.calibration import CalibratedClassifierCV

pipeline = logisticPipelineImportance()
pipeline.set_params(**searchLogistic.best_params_)

pipeline = CalibratedClassifierCV(estimator=pipeline, cv=10)
pipeline.fit(train[logisticFeature], train[label])

train_score = round(pipeline.score(train[logisticFeature], train[label]), 3)
valid_score = round(pipeline.score(validation[logisticFeature], validation[label]), 3)

print(f"train score: {train_score:.3f}")
print(f"valid score: {valid_score:.3f}")
```

```python
train_score = pipeline.score(train[logisticFeature], train[label])
test_score = pipeline.score(test[logisticFeature], test[label])

print(f"Train score: %.3f" % train_score )
print(f"Test score: %.3f" % test_score)
```

```python
from scipy.stats import randint, uniform

param_distributions = {
    'n_estimators': randint(50, 200),  # จำนวน base estimators
    'learning_rate': uniform(0.01, 1.0),  # อัตราการเรียนรู้
    'algorithm': ['SAMME', 'SAMME.R'],  # อัลกอริธึมที่ใช้
    'estimator__max_depth': randint(1, 20),  # ความลึกสูงสุดของต้นไม้
    'estimator__min_samples_split': randint(2, 20),  # จำนวนตัวอย่างขั้นต่ำในการแบ่งปันโหนด
    'estimator__min_samples_leaf': randint(1, 20),  # จำนวนตัวอย่างขั้นต่ำในโหนดใบไม้
    'estimator__max_features': randint(1, 10)  # จำนวนฟีเจอร์สูงสุดที่ใช้ในการแบ่งปันโหนด
}

random_search = RandomizedSearchCV(
    Adatree,
    param_distributions,
    n_iter=100,  # จำนวนชุดพารามิเตอร์ที่ต้องการทดลอง
    cv=5,  # จำนวนรอบในการ cross-validation
    verbose=1,  # ระดับการแสดงผล
    random_state=42,  # ตั้งค่า seed สำหรับการสุ่ม
    n_jobs=-1  # ใช้ CPU ทั้งหมดที่มีอยู่
)

# ฝึก RandomizedSearchCV
random_search.fit(train[feature], train[label])

# ดูพารามิเตอร์ที่ดีที่สุด
print("Best parameters:", random_search.best_params_)

# ดูคะแนนที่ดีที่สุด
print("Best score:", random_search.best_score_)
```

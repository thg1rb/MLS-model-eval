```python
import pandas as pd

df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv")
feature = df.drop(columns="Diabetes_binary").columns
label = "Diabetes_binary"
df.drop_duplicates(inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler

train, test = train_test_split(df, test_size = 0.3, stratify = df.Diabetes_binary, random_state = 1123)
resampler = RandomUnderSampler(random_state=1909)
# train[feature], train[label] = resampler.fit_resample(train[feature], train[label])
train.dropna(inplace=True)
test, validation = train_test_split(test, test_size = 0.3, stratify = test.Diabetes_binary, random_state = 1002)
# train1, train = train_test_split(train, test_size=0.7, stratify = train.Diabetes_binary, random_state = 1123)
len(train), len(test), len(validation), #len(train1)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

features = ["GenHlth", "BMI", "Age", "Fruits", "Smoker", "NoDocbcCost"]

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(train[features], train[label])

# Train a RandomForestClassifier with class weights
model = RandomForestClassifier(class_weight='balanced', random_state=42)
model.fit(X_train_resampled, y_train_resampled)

# y_pred = model.predict(test[feature])
# print(classification_report(test[label], y_pred))
```

```python
model.score(train[features], train[label])
```

```python
model.score(validation[features], validation[label])
```

```python
model.score(test[features], test[label])
```
